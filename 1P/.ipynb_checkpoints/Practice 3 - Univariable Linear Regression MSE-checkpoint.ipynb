```python
import pandas as pd
import numpy as np
import copy
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline
```

```python
# Setting configuration for pandas dataframes visualizations
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

```python
# Create a data.frame with the file given
df = pd.read_csv('./dataset_ejercicio_I_regresion_lineal.csv', sep=',', engine='python')
```

```python
df.head(10)
```

```python
# Indicating tag column name
tag_name = 'price'
```

```python
# Corpus without tags
X = df.drop(tag_name, axis=1).values
X
```

```python
# Tags
y = df[tag_name].values
y
```

```python
# Getting training and test data
X_e, X_p, y_e, y_p = train_test_split(X, y, test_size = 0.1, shuffle = True, random_state = 0)
```

```python
# Defining MSE function and print generated line function
def F(w, X, y):
    return sum((w * x - y)**2 for x, y in zip(X, y))/len(y)


def dF(w, X, y):
    return sum(2*(w * x - y) * x for x, y in zip(X, y))/len(y)


def print_line(points, w, iteration, ax1, line_color = None, line_style = 'dotted'):
    list_x = []
    list_y = []
    for index, tuple in enumerate(points):
        x = tuple[0]
        y = x * w
        list_x.append(x)
        list_y.append(y)
    ax1.text(x,y, iteration, horizontalalignment='right')
    ax1.plot(list_x, list_y, color = line_color, linestyle= line_style)

```

```python
# Defining the process to generate the linear regression by MSE
def plot_uni_linear_regression_mse(X, y, alpha, iterations):
    list_error = []
    list_w = []

    fig = plt.figure(figsize=(15, 5))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.set_title("Linear regression")
    ax1.set(xlabel="size", ylabel="price")
    ax2 = fig.add_subplot(1, 2, 2)
    ax2.set_title("Loss function")
    ax2.set(xlabel="weight", ylabel="error")
    ax1.scatter(X, y)
    
    print("\n\n\033[1malpha =", str(alpha), "\titerations =", str(iterations), "\033[0m")
    last_iterations = 5
    print("\033[94mLast", str(last_iterations), "iterations:\033[94m")
    
    w = 0
    for t in range(iterations):
        error = F(w, X, y)
        gradient = dF(w, X, y)
        
        ax2.scatter(w, error)
        ax2.text(w, error, t, horizontalalignment='right')
        list_w.append(w)
        list_error.append(error)

        w = w - alpha * gradient
        if t + 1 > iterations - last_iterations:
            print ('gradient = {}'.format(gradient))
            print ('iteration {}: w = {}, F(w) = {}'.format(t, w, error))
        print_line(zip(X, y), w, t, ax1)

    print_line(zip(X, y), w, t, ax1, 'red', 'solid')
    ax2.plot(list_w, list_error, color = 'red', linestyle = 'solid')

    plt.show()
    
    return w, error
```

```python
# Defining multiple values to generate different linear regressions
alphas = [0.00001, 0.0001]
iterations = [10, 50, 100]
w_all_combinations = dict()
mse_training_data = dict()
```

```python
# Make all combinations with the above values
for a in alphas:
    for i in iterations:
        w_all_combinations[(a, i)], mse_training_data[(a, i)] = plot_uni_linear_regression_mse(copy.deepcopy(X_e), copy.deepcopy(y_e), a, i)
```

```python
w_all_combinations
```

```python
mse_training_data
```

```python
# Defining function to predict the wanted value
def predict(w, X):
    res = list()
    X = [x[0] for x in X]
    for i in range(len(X)):
        res.append((w * X[i])[0])
    return res
```

```python
# Selecting parameters that had a good performance while generating linear regression
alpha = 0.00001
itera = 100
```

```python
# Plotting the linear regression by training data with the selected parameters
plt.scatter(X_e, y_e)
plt.title("Fit linear regression by training data")
plt.plot(X_e, predict(w_all_combinations[alpha, itera], X_e), color = "green")
```

```python
# Predicted values with test data
res = predict(w_all_combinations[(alpha, itera)], X_p)
res
```

```python
# Plotting the linear regression to test data with the selected parameters
plt.scatter(X_p, y_p)
plt.title("Linear regression to test data")
plt.plot(X_p, predict(w_all_combinations[alpha, itera], X_p), color = "green")
```

```python
# Calculating the MSE of training data corresponding to the selected parameters
mse_training_data[(alpha, itera)]
```

```python
# Calculating the MSE of test data corresponding to the selected parameters
mse_test_data = F(w_all_combinations[(alpha, itera)], X_p, y_p)
mse_test_data
```
