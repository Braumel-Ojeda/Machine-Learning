Using the heart.csv dataset do the following:

```python
# Import the needed libraries
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay 
from tabulate import tabulate
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get data
df = pd.read_csv('./heart.csv', sep = ',', engine = 'python')
```

```python
df.head(10)
```

```python
# Indicating tag column name
tag_name = 'target'
```

```python
# Corpus without tags
X = df.drop(tag_name, axis = 1).values
X
```

```python
# Tags
y = df[tag_name].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = True, random_state = 0)
```

```python
# Simple implementation of SVM
class SVM:
    
    def fit(self, X_train, y_train):
        pos_data = list()
        neg_data = list()
        for i in range(len(y_train)):
            if y_train[i] == 1:
                pos_data.append(X_train[i])
            else:
                neg_data.append(X_train[i])
        pos_data = np.array(pos_data)
        neg_data = np.array(neg_data)
        
        self.c_pos = np.mean(pos_data, axis = 0)
        self.c_neg = np.mean(neg_data, axis = 0)
        self.c = np.array(self.c_pos + self.c_neg) / 2
        self.c_norm = np.linalg.norm(self.c)
    
    def predict(self, X_test):
        y_predict = list()
        for x in X_test:
            projection = np.dot(x, self.c) / self.c_norm
            if projection > self.c_norm:
                y_predict.append(1)
            else:
                y_predict.append(0)
        
        return y_predict
```

```python
svm = SVM()
svm.fit(X_train, y_train)
```

```python
print('\033[1mPositive c vector\033[0m\n', svm.c_pos)
```

```python
print('\033[1mNegative c vector\033[0m\n', svm.c_neg)
```

```python
print('\033[1mc vector\033[0m\n', svm.c)
```

```python
print('\033[1mc vector norm\033[0m\n', svm.c_norm)
```

```python
y_predict = svm.predict(X_test)
```

```python
results = list()
for i in range(len(y_test)):
    results.append([y_test[i], y_predict[i]])

print(tabulate(results, headers = ['y original', 'y predicted'], tablefmt = 'fancy_grid', numalign = 'center',))
```

```python
target_names = ['0', '1']
```

```python
print(classification_report(y_test, y_predict, target_names = target_names))
```

```python
cm = confusion_matrix(y_test, y_predict, labels = [0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = target_names)
disp.plot()
plt.show()
```
